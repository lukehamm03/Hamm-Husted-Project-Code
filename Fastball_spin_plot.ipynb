```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
import pickle

file_path = r"C:\Users\lukeh\Downloads\refined_data.csv"
data = pd.read_csv(file_path)

features = [
    'fastball_avg_spin', 'offspeed_avg_break', 'fastball_avg_break_z_induced',
    'fastball_avg_speed', 'breaking_avg_speed', 'arm_angle', 'offspeed_avg_speed'
]
X = data[features]
y = data['elbow?']  

X = X.fillna(X.mean())

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

model = LogisticRegression(max_iter=500)
model.fit(X_train_scaled, y_train)

with open("logistic_model.pkl", "wb") as f:
    pickle.dump(model, f)

with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)

print("New model and scaler saved successfully.")
```

```python
import numpy as np
import pickle

with open("logistic_model.pkl", "rb") as f:
    model = pickle.load(f)
with open("scaler.pkl", "rb") as f:
    scaler = pickle.load(f)

feature_ranges = {
    "fastball_avg_spin": np.linspace(1500, 3000, 50),
    "offspeed_avg_break": np.linspace(0, 20, 50),
    "fastball_avg_break_z_induced": np.linspace(0, 30, 50),
    "fastball_avg_speed": np.linspace(80, 105, 50),
    "breaking_avg_speed": np.linspace(70, 95, 50),
    "arm_angle": np.linspace(30, 90, 50),
    "offspeed_avg_speed": np.linspace(60, 85, 50),
}

baseline_values = {
    "fastball_avg_spin": 2200,
    "offspeed_avg_break": 12,
    "fastball_avg_break_z_induced": 15,
    "fastball_avg_speed": 95,
    "breaking_avg_speed": 85,
    "arm_angle": 50,
    "offspeed_avg_speed": 80,
}

def calculate_percent_change(model, scaler, feature_name, feature_range):
    percent_changes = []
    for value in feature_range:
        input_features = baseline_values.copy()
        input_features[feature_name] = value
        features = np.array([[input_features[key] for key in baseline_values]])
        
        scaled_features = scaler.transform(features)
        risk = model.predict_proba(scaled_features)[0][1] * 100
        percent_changes.append(risk)
    
    return percent_changes
```

```python
import plotly.express as px
import plotly.io as pio

def create_interactive_plot(feature_name):
    feature_range = feature_ranges[feature_name]
    percent_changes = calculate_percent_change(model, scaler, feature_name, feature_range)
    
    fig = px.line(
        x=feature_range,
        y=percent_changes,
        title=f"Impact of {feature_name.replace('_', ' ').title()} on Injury Risk",
        labels={
            "x": f"{feature_name.replace('_', ' ').title()} (Log Scale)",
            "y": "Percent Change in Injury Risk (%) (Log Scale)"
        },
    )
    fig.update_xaxes(type="log", title_text=f"{feature_name.replace('_', ' ').title()} (Log Scale)")
    fig.update_yaxes(type="log", title_text="Percent Change in Injury Risk (%) (Log Scale)")
    fig.update_layout(
        title_font=dict(size=20),
        title_x=0.5,
        margin=dict(l=40, r=40, t=40, b=40),
        font=dict(size=14),
    )
    return fig

fig = create_interactive_plot("fastball_avg_spin")

output_html_file = "fastball_avg_spin_injury_risk.html"
pio.write_html(fig, file=output_html_file, auto_open=False)

print(f"HTML visualization saved as: {output_html_file}")
```
